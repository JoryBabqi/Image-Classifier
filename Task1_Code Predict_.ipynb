```python
import tensorflow as tf  # Import TensorFlow library
from tensorflow import keras  # Import Keras from TensorFlow
import numpy as np  # Import NumPy for numerical operations
from PIL import Image  # Import PIL for image processing

# Load the pre-trained Keras model from a file
model = keras.models.load_model('keras_model.h5')

def load_image(image_path):
    """
    Load and preprocess an image for prediction.

    Args:
        image_path (str): The path to the image file.

    Returns:
        np.ndarray: A batch of images with the shape (1, 224, 224, 3).
    """
    img = Image.open(image_path)  # Open the image file
    img = img.resize((224, 224))  # Resize image to 224x224 pixels
    img_array = np.array(img) / 255.0  # Convert image to array and normalize pixel values
    return np.expand_dims(img_array, axis=0)  # Add a new dimension for batch size

def predict(image_path):
    """
    Predict the class of an image using the model.

    Args:
        image_path (str): The path to the image file.

    Returns:
        np.ndarray: The predicted class index.
    """
    image = load_image(image_path)  # Load and preprocess the image
    predictions = model.predict(image)  # Get model predictions for the image
    predicted_class = np.argmax(predictions, axis=1)  # Get the index of the highest prediction score
    return predicted_class  # Return the predicted class index

# List of image paths to predict
image_paths = ['smallcat.png', 'bigcat.png', 'dog.png', 'dog1.png']
predictions = {}  # Initialize an empty dictionary to store predictions

# Loop through each image path and make predictions
for image_path in image_paths:
    predicted_class = predict(image_path)  # Get the predicted class for the current image
    predictions[image_path] = predicted_class  # Store the result in the predictions dictionary

# Print the predictions for each image
for img, pred in predictions.items():
    print(f'Predicted class for {img}: {pred}')  # Display the predicted class for each image
```
